```python
import pandas as pd

movies = pd.read_csv(r"C:\Users\hunte\Documents\Analyst Workshop\Pandas_For_Data_Analysis\11. Real World Data Analysis Projects\Movie Genres\imdb_movies.csv")
```

```python
pd.set_option('display.max.rows', 11000)
pd.set_option('display.max.columns', 22)
```

```python
movies.info()
```

```python
movies.head()
```

```python
movies[movies.duplicated()]
```

```python
movies.drop_duplicates(inplace=True)
```

```python
movies[movies.duplicated()]
```

```python
movies.dropna(subset=['genres'], inplace=True)
```

```python
movies['profit'] = movies['revenue']-movies['budget']
```

```python
movie_genre = movies[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
```

```python
movie_genre.head()
```

```python
from pandas import Series, DataFrame
```

```python
split = movie_genre['genres'].str.split('|').apply(Series, 1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movie_genre['genres']
movie_genre = movie_genre.join(split)
```

```python
movie_genre
```

### Research Questions (Q): 

1. Which genres are the most common (number of movies made)?

Answer: The genres that are most common are Drama, Comedy, Thriller, and Action

```python
genres_count = pd.DataFrame(movie_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending = True)
```

```python
genres_count
```

```python
genres_count['original_title'].plot.pie(title = 'Movies Per Genre in %', autopct = '%1.1f', figsize = (10,10))
```

```python
genres_count['original_title'].plot.barh(title = 'Movies Per Genre', color = 'Darkblue', figsize = (10,9))
```

2. Which genres have high avg. budget and revenue?

Answer: The genres that have high avg. budget and revenue are Adventure, Fantasy, and Action

```python
genres_avg = movie_genre.groupby('genres_split').mean(numeric_only=True)
pd.options.display.float_format = '{:2f}'.format
genres_avg
```

```python
genres_avg.sort_values('budget', ascending = True, inplace = True)
```

```python
genres_avg[['budget', 'revenue']].plot.barh(title = 'Budget and Revenue by Genre', color = ('Darkblue','c'), figsize = (10,9))
```

3. Which genres have high avg. profit

Answer: The genres that have high avg. profit are Adventure, Fantasy, and Animation

```python
genres_avg.sort_values('profit', ascending = True, inplace = True)
```

```python
genres_avg[['profit']].plot.barh(title = 'Profit by Genre', color = ('Darkblue','c'), figsize = (10,9))
```

4. Which genres have high avg. popularity?

Answer: The genres that have high avg. popularity are Adventure, Science Fiction, and Fantasy.

```python
genres_avg.sort_values('popularity', ascending = True, inplace = True)
```

```python
genres_avg[['popularity']].plot.barh(title = 'Popularity by Genre', color = ('Darkblue','c'), figsize = (10,9))
```

5. Which genres have highest number of movies with a voting avg. >=8?

Answer: The genres that have the highest number of movies with a voting avg. >=8 are Drama, Crime, and Documentary.

```python
movie_genre.head()
```

```python
voting_avg = movie_genre[(movie_genre['vote_average'] >= 8) & (movie_genre['vote_count'] >= 50)]
voting_avg.head()
```

```python
genres_vote = voting_avg.groupby('genres_split').size().sort_values(ascending=True)
pd.DataFrame(genres_vote)
```

```python
genres_vote.plot.barh(title = 'Vote Average by Genre', color = 'DarkBlue',figsize = (10,9))
```

### Research Hypotheses (H): 

1. The best movies according to vote avg. return high profit and revenue.

The hypothesis was slightly off as the correlation was quite slim

```python
movies.drop_duplicates(inplace = True)
movies['profit'] = movies['revenue'] - movies['budget']
movies_genre = movies[['popularity','budget','revenue','original_title','runtime','genres','release_date', 'release_year','vote_count','vote_average','profit']]
```

```python
movies_genre.head()
```

```python
movies_counted = movies_genre[movies_genre['vote_count']>= 50]
movies_counted.corr(method='spearman', numeric_only= True)
```

```python
movies_counted.plot.scatter(x = 'vote_average',y = 'profit')
```

```python
import seaborn as sns
```

```python
sns.regplot(x = 'vote_average', y = 'profit', data = movies_counted, line_kws = {'color':'red'})
```

```python
sns.regplot(x = 'vote_average', y = 'revenue', data = movies_counted, line_kws = {'color':'red'})
```

2. The best movies according to popularity return high profit and revenue.

This hypothesis was correct and it shows strongly in the scatter plot below

```python
movies_counted.corr(method='spearman', numeric_only = True)
```

```python
import matplotlib.pyplot as plt
sns.regplot(x = 'popularity', y = 'revenue', data = movies_counted, line_kws = {'color':'red'})
plt.figure(figsize = (10,15))
plt.show()
```

3. Highly budgeted movies return high revenue and profit.

This hypothesis was accurate. The scatterplot below shows a healthy correlation.

```python
movies_counted.head()
```

```python
sns.regplot(x = 'budget', y = 'profit', data = movies_counted, line_kws = {'color':'red'})
plt.figure(figsize = (10,15))
plt.show()
```

4. Highly budgeted movies have a high popularity.

As the budget increases, it does not increase its popularity that much. 
There are some outliers in the data that may compliment the hypothesis in terms of the correlation

```python
sns.regplot(x = 'budget', y = 'popularity', data = movies_counted, line_kws = {'color':'red'})
plt.figure(figsize = (10,15))
plt.show()
```
